1. Можно ли отобрать наиболее значимые признаки с помощью PCA? Ответ объясните.

Главные компоненты PCA несут информацию о доле изменчивости признаков: первая имеет данные о наибольшей доле, вторая - о наибольшей оставшейся доле и т.д. Т.е. по матрице весов, полученной в PCA преобразовании, можно оценить вклад каждого признака в главные компоненты, учитывая также их долю. При этом мне видится только возможность определения наименее значимых признаков по их меньшему вкладу, а вот определить/ранжировать признаки по их важности на основе вклада в главные компоненты однозначно не получится. В задаче №3 протестирована данная логика. Также там рассмотрен анализ только в рамках вклада в первую главную компоненту.
